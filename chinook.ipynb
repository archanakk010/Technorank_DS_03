## 6.Filtering Data

##### 1.Customers from the USA

```python
query = """
SELECT FirstName, LastName, Country
FROM Customer
WHERE Country = 'USA'
"""

df = pd.read_sql_query(query, conn)
df

```

##### 2.Sales Greater Than $10

```python
query = """
SELECT InvoiceId, CustomerId, Total
FROM Invoice
WHERE Total > 10
"""

df = pd.read_sql_query(query, conn)
df

```

## 7.Sorting Examples

##### 1.Sort Customers by Last Name

```python
query = """
SELECT FirstName, LastName, Country
FROM Customer
ORDER BY LastName ASC
"""
pd.read_sql_query(query, conn)
```

##### 2.Sort Invoices by Total Amount (Descending)

```python
query = """
SELECT InvoiceId, CustomerId, Total
FROM Invoice
ORDER BY Total DESC
"""
pd.read_sql_query(query, conn)
```

##  8.Aggregation Queries

##### 1.Total Sales per Country

```python

query = """
SELECT BillingCountry, SUM(Total) AS TotalSales
FROM Invoice
GROUP BY BillingCountry
ORDER BY TotalSales DESC
"""

df = pd.read_sql_query(query, conn)
df


```

##### 2.Number of Customers per Country

```python

query = """
SELECT Country, COUNT(*) AS NumCustomers
FROM Customer
GROUP BY Country
ORDER BY NumCustomers DESC
"""

df = pd.read_sql_query(query, conn)
df
```

##  Joining Tables

##### 1.Total Spent by Each Customer

```python
query = """
SELECT c.FirstName || ' ' || c.LastName AS CustomerName, SUM(i.Total) AS TotalSpent
FROM Customer c
JOIN Invoice i ON c.CustomerId = i.CustomerId
GROUP BY c.CustomerId
ORDER BY TotalSpent DESC
LIMIT 5
"""
pd.read_sql_query(query, conn)
```

#####  2.Join Customer and Invoice to View Sales Per Customer

```python

query = """
SELECT c.FirstName || ' ' || c.LastName AS CustomerName, i.Total, i.InvoiceDate
FROM Customer c
JOIN Invoice i ON c.CustomerId = i.CustomerId
ORDER BY i.Total DESC
"""

df = pd.read_sql_query(query, conn)
df

```

##### 3.Top 5 Customers by Total Purchase

```python

query = """
SELECT c.FirstName || ' ' || c.LastName AS CustomerName, SUM(i.Total) AS TotalSpent
FROM Customer c
JOIN Invoice i ON c.CustomerId = i.CustomerId
GROUP BY c.CustomerId
ORDER BY TotalSpent DESC
LIMIT 5
"""

df = pd.read_sql_query(query, conn)
df

```
